```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
flights = sns.load_dataset('flights')

flights.head()
```

```python
titanic = pd.read_csv('Titanic.csv')
titanic.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

**BIVARIATE Analysis**

Scatter plot (Numerical-Numerical)

```python
#two columns and both are numerical and we are using Tips Dataset
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
# Multivariate Analysis is all about doing analysis on more than 2columns
sns.scatterplot(x='total_bill',y='tip',data=tips, hue = tips['sex'], style = tips['smoker'], size=tips['size'])
```

***Bar Plot ***

Numerical - Categorical

```python
#finding the average age of the Pclass in the titanic dataset
sns.barplot(x='Pclass',y='Fare',data=titanic, hue=titanic['Sex'])
```

**Box Plot**

Numerical - Categorical

```python
sns.boxplot(x='Sex',y='Age',data=titanic, hue=titanic['Survived'])
```

**Distplot**

Numerical - Categorical

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

**HeatMap**

Categorical - Categorical

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='bar')
```

**Cluster Map**

Categorical - Categorical

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

**PairPlot**

```python
#collection of scatterplot i.e let say you have multiple numerical column and for each numerical column you have to plot scatterplot against each and every other numerical column. So pairplot solve this problem by ploting the scatterplot of every column with each other
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

**Lineplot**

```python
#lineplot is a special case and if we join the points of scatterplot then we call it as Lineplot. We use this when we have quantity on x-axis in time based form eg. time, month, year
flights.head()
```

```python
#how many passengers travelled each year
flights.groupby('year')['passengers'].sum().reset_index()
```

```python
sns.lineplot(x='year',y='passengers',data=flights)
```

```python
flights.head()
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
